```python
from intent_classification import intent_classification_system_message

call_llm = get_groq_client(
    system_prompt=intent_classification_system_message,
    user_input=user_message_template,
    model="openai/gpt-oss-120b",
    temperature=0.1
)
print(call_llm)
```

```python
output_schema = {
 "blog_intent": "",
 "target_reader_maturity": "BEGINNER | INTERMEDIATE | ADVANCED",
 "recommended_content_angle": "",
 "reasoning_summary": ""
}
```

```python
import json
from jsonschema import validate, ValidationError
```

```python

try:
    parsed_output = json.loads(call_llm)

    validate(instance=parsed_output, schema=output_schema)

    print("VALID JSON OUTPUT:\n")
    print(json.dumps(parsed_output, indent=2))

except json.JSONDecodeError:
    print("ERROR: Model output is not valid JSON.")

except ValidationError as e:
    print("SCHEMA VALIDATION ERROR:")
    print(e)
```

```python
import os
os.makedirs("outputs", exist_ok=True)

# File path
output_path = "outputs/blog_intent.json"

# Save JSON output
with open(output_path, "w") as f:
    json.dump(parsed_output, f, indent=2)

print(f"Output successfully saved to: {output_path}")
```
