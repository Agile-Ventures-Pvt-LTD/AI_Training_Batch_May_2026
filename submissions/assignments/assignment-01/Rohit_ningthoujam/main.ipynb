# SELF CONSISTENCY


```python
from groq_client import run_case, save_output
from prompts import SELF_CONSISTENCY_CASE_1_PROMPT
from collections import Counter
import json
```

```python
results = []

for i in range(5):
    result = run_case(
        f"case_5_1_run_{i+1}",
        SELF_CONSISTENCY_CASE_1_PROMPT,
        temperature=0.7
    )
    results.append(result)
    results
```

```python
def extract_value(x):
    try:
        # case: list output (bad model behavior)
        if isinstance(x, list):
            for v in x:
                try:
                    return float(v)
                except:
                    continue

        # case: dict output
        if isinstance(x, dict):
            if "reimbursable_amount" in x:
                return float(x["reimbursable_amount"])
            return None

        # case: direct number
        return float(x)

    except:
        return None
```

```python
values = [extract_value(r) for r in results]
values = [v for v in values if v is not None]
```

```python

counts = Counter(values)
```

```python
final_amount = counts.most_common(1)[0][0] if counts else 0
```

```python
output = {
    "individual_answers": results,
    "final_reimbursable_amount": final_amount,
    "consistency_count": dict(counts),
    "final_decision": f"Selected most frequent value: {final_amount}",
    "reasoning_summary": "Self-consistency applied using 5 independent API calls with majority voting."
}
```

```python
save_output(
    "self_consistency_case_5_1_output.json",
    output
)
```

```python
print(json.dumps(output, indent=2))
```

# Case - 2 

```python
from groq_client import run_case, save_output
from prompts import SELF_CONSISTENCY_PROMPT_CASE_5_2
from collections import Counter
import json

def extract_label(response):
    if isinstance(response, dict):
        if "raw_response" in response:
            return str(response["raw_response"]).strip().upper()
        if "risk_level" in response:
            return str(response["risk_level"]).strip().upper()
        return str(response)

    return str(response).strip().upper()

runs = []

for i in range(5):
    result = run_case(
        f"self_consistency_case_5_2_run_{i+1}",
        SELF_CONSISTENCY_PROMPT_CASE_5_2
    )

    label = extract_label(result)

    runs.append({
        "run": i + 1,
        "raw_response": result,
        "parsed": label
    })

valid_votes = [
    r["parsed"]
    for r in runs
    if r["parsed"] and r["parsed"] not in ["ERROR", "INVALID JSON RETURNED BY MODEL"]
]

vote_count = Counter(valid_votes)

final_risk = vote_count.most_common(1)[0][0] if vote_count else "UNKNOWN"

output = {
    "runs": runs,
    "risk_level_votes": dict(vote_count),
    "final_risk_level": final_risk,
    "disagreement_analysis": "Differences arise due to interpretation of VPN exclusion and file download threshold rules.",
    "final_reasoning_summary": "Self-consistency applied across 5 independent runs; majority vote used for final risk classification."
}

save_output(
    "self_consistency_case_5_2_output.json",
    output
)

print(json.dumps(output, indent=2))
```
