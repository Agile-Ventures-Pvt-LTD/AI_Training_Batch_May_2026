```python
import importlib
import groq_client

importlib.reload(groq_client)

raw_output = groq_client.call_llm(
    system_prompt=prompt,
    user_input=vendor_text,
    model="openai/gpt-oss-120b",
    temperature=0.1
)
```

```python

try:
    parsed_output = json.loads(raw_output)

    validate(instance=parsed_output, schema=schema)

    print("VALID JSON OUTPUT:\n")
    print(json.dumps(parsed_output, indent=2))

except json.JSONDecodeError:
    print("ERROR: Model output is not valid JSON.")

except ValidationError as e:
    print("SCHEMA VALIDATION ERROR:")
    print(e)
```
