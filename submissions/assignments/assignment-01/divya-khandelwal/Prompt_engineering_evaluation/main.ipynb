### Self Consistency 

```python
import prompts
from groq_client import run_case, save_output
```

```python
responses = []

for i in range(5):

    result = run_case(
        system_message=prompts.SYSTEM_MESSAGE_self_consistency_case1,
        user_message=prompts.USER_MESSAGE_self_consistency_case1,
        temperature=0.7
    )

    responses.append(result)

print(responses)
```

```python
amounts = []

for response in responses:

    amount = response.get("reimbursable_amount")

    amounts.append(amount)

print(amounts)
```

```python
from collections import Counter

consistency_count = Counter(amounts)

print(consistency_count)
```

```python
final_amount = consistency_count.most_common(1)[0][0]

print(final_amount)
```

```python
final_output_self_consistency_case1 = {

    "individual_answers": amounts,

    "final_reimbursable_amount": final_amount,

    "consistency_count": dict(consistency_count),

    "final_decision": f"Most consistent reimbursable amount is {final_amount}",

    "reasoning_summary": "Multiple independent reasoning runs were generated and the most consistent reimbursement amount was selected."

}

print(final_output_self_consistency_case1)
```

```python
save_output(
    "outputs/self_consistency_outputs.json",
    final_output_self_consistency_case1
)
```

### Self Consistency Case-2

```python
import prompts
from groq_client import run_case, save_output
```

```python
risk_runs = []

for i in range(5):

    result = run_case(
        system_message=prompts.SYSTEM_MESSAGE_self_consistency_case2,
        user_message=prompts.USER_MESSAGE_self_consistency_case2,
        temperature=0.7
    )

    risk_runs.append(result)

print(risk_runs)
```

```python
risk_levels = []

for run in risk_runs:

    level = run.get("risk_level")

    risk_levels.append(level)

print(risk_levels)
```

```python
from collections import Counter

risk_level_votes = Counter(risk_levels)

print(risk_level_votes)
```

```python
final_risk_level = risk_level_votes.most_common(1)[0][0]

print(final_risk_level)
```

```python
final_output_self_consistency_case2 = {

    "runs": risk_levels,

    "risk_level_votes": dict(risk_level_votes),

    "final_risk_level": final_risk_level,

    "disagreement_analysis": "Some runs may incorrectly classify HIGH risk by focusing only on file downloads without validating the new-country condition.",

    "final_reasoning_summary": "The most consistent result was selected after multiple independent reasoning runs."

}

print(final_output_self_consistency_case2)
```

```python
save_output(
    "outputs/self_consistency_outputs.json",
    final_output_self_consistency_case2
)
```
