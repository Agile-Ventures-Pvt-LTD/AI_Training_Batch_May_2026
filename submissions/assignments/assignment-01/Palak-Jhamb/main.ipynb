# Assignment-1 Foundation of Generative AI

Case 1.1 - Zero-Shot Risk Classification for Vendor Onboarding


```python
import json
from pathlib import Path
from prompts import SYSTEM_PROMPT_VENDOR_RISK, USER_PROMPT_VENDOR_RISK
from groq_client import GroqClient
```

```python
client = GroqClient()
```

```python
raw_response = client.generate(
    system_prompt=SYSTEM_PROMPT_VENDOR_RISK,
    user_prompt=USER_PROMPT_VENDOR_RISK
)

print(raw_response)
```

```python
parsed_output = json.loads(raw_response)

parsed_output
```

```python
output_dir = Path("outputs")
output_dir.mkdir(exist_ok=True)

output_file = output_dir / "zero_shot_outputs_case1.json"

with open(output_file, "w") as f:
    json.dump(parsed_output, f, indent=4)

print(f"Saved output to: {output_file}")
```
