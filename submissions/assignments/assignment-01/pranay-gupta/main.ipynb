```python
import os
import json
```

```python
# zero-shot template
from groq_client import client
from prompts import zeroshot_system_message, zeroshot_user_input
def zeroshot_call_llm(zeroshot_system_message, zeroshot_user_input, model="openai/gpt-oss-120b", temperature=0.3):
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "system", "content":zeroshot_system_message},{"role": "user", "content": zeroshot_user_input}],
        temperature=temperature,
    )
    return response.choices[0].message.content

print(zeroshot_call_llm(zeroshot_system_message, zeroshot_user_input))

```

```python
OUTPUT_FOLDER = "outputs"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)


def save_json_output(filename="zeroshot_output1.json"):
    """
    Save LLM JSON response into outputs folder.
    """
    llm_output = zeroshot_call_llm(
    zeroshot_system_message,
    zeroshot_user_input
    )
    try:
        # Convert string response into JSON
        json_data = json.loads(llm_output)

        # File path
        file_path = os.path.join(OUTPUT_FOLDER, filename)

        # Save JSON file
        with open(file_path, "w", encoding="utf-8") as file:
            json.dump(
                json_data,
                file,
                indent=4,
                ensure_ascii=False
            )

        print(f"JSON saved successfully at: {file_path}")

    except json.JSONDecodeError:
        print("Invalid JSON response from LLM")


save_json_output()
```

```python
from prompts import (
    fewshot1_system_message,
    assistant1_output_examples,
    fewshot1_user_input
)
from groq_client import client

def fewshot1_call_llm(fewshot1_system_message, assistant1_output_examples, fewshot1_user_input, model="openai/gpt-oss-120b", temperature=0.3):
    messages = [
        {
            "role": "system",
            "content": fewshot1_system_message
        }
    ]
    messages.extend(assistant1_output_examples)
    messages.append(
        {
            "role": "user",
            "content": fewshot1_user_input
        }
    )

    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,
    )
    return response.choices[0].message.content

print(fewshot1_call_llm(fewshot1_system_message, assistant1_output_examples, fewshot1_user_input))
```

## Self-Consistency

#### Case 5.1 - Self-Consistency for Complex Policy Interpretation

```python

from groq_client import client

from prompts import (
    self1_system_message,
    self1_answers_template,
    self1_consistency_template,
    self1_claim_context
)

answers_prompt = [
    {
        "role": "system",
        "content": self1_system_message
    },
    {
        "role": "user",
        "content": self1_answers_template.format(
            context=self1_claim_context,
            question="What is the final reimbursable amount?",
            num_answers=5
        )
    }
]

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=answers_prompt,
    temperature=0.2
)

self_consistency_answers = response.choices[0].message.content

print("STEP 1 OUTPUT:\n")
print(self_consistency_answers)
```

```python
consistency_prompt = [
    {
        "role": "system",
        "content": self1_system_message
    },
    {
        "role": "user",
        "content": self1_consistency_template.format(
            question="What is the final reimbursable amount?",
            answers=self_consistency_answers
        )
    }
]

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=consistency_prompt,
    temperature=0
)

final_result = response.choices[0].message.content

print("\nFINAL OUTPUT:\n")
print(final_result)
```

```python
import os
import json

OUTPUT_FOLDER = "outputs"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)


def save_json_output_clean(llm_output, filename="self_output1.json"):

    try:
        if isinstance(llm_output, str):
            data = json.loads(llm_output)
        else:
            data = llm_output

        file_path = os.path.join(os.getcwd(), OUTPUT_FOLDER, filename)

        with open(file_path, "w", encoding="utf-8") as f:
            json.dump(
                data,
                f,
                indent=4,
                ensure_ascii=False
            )

        print("✅ Saved clean JSON at:", file_path)

    except Exception as e:
        print("❌ Error while saving JSON:")
        print(e)

save_json_output_clean(final_result)
```
