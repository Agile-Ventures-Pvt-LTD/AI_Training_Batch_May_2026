## Function for five independent model calls Case 5.1. 

```python
OUTPUT_DIR = "outputs"

def run_self_consistency_case(system_prompt, user_prompt, case_name, n=5):
    runs = []
    
    # Run multiple independent calls
    for i in range(n):
        response_text = call_llm(system_prompt, user_prompt, temperature=0.7)  # higher temp for diversity
        result = extract_json(response_text)
        if result:
            runs.append(result)
    
    # Decide schema type based on keys in first run
    if not runs:
        return None
    
    first = runs[0]
    
    if "final_reimbursable_amount" in first:  # Case 5.1
        amounts = [ans.get("final_reimbursable_amount") for ans in runs if ans]
        counter = Counter(amounts)
        final_amount, count = counter.most_common(1)[0] if counter else (0, 0)
        
        final_output = {
            "individual_answers": runs,
            "final_reimbursable_amount": final_amount,
            "consistency_count": dict(counter),
            "final_decision": "APPROVE" if final_amount > 0 else "REJECT",
            "reasoning_summary": f"Selected the majority reimbursable amount ({final_amount}) from {n} independent runs."
        }
    
    elif "final_risk_level" in first:  # Case 5.2
        levels = [ans.get("final_risk_level") for ans in runs if ans]
        counter = Counter(levels)
        final_level, count = counter.most_common(1)[0] if counter else ("", 0)
        
        final_output = {
            "runs": runs,
            "risk_level_votes": dict(counter),
            "final_risk_level": final_level,
            "disagreement_analysis": "Some runs may misapply rules (e.g., flagging HIGH risk for downloads alone). Majority vote selected.",
            "final_reasoning_summary": f"Selected the majority risk classification ({final_level}) from {n} independent runs."
        }
    
    else:
        raise ValueError("Unknown schema type in model output.")
    
    # Save output
    if not os.path.exists(OUTPUT_DIR):
        os.makedirs(OUTPUT_DIR)
    save_output(os.path.join(OUTPUT_DIR, f"{case_name}.json"), final_output)
    
    return final_output
```
