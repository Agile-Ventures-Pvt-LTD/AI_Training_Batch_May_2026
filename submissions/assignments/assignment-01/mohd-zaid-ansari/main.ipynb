ZERO-SHOT PROMPTING CASE - 1.1

```python
from groq_client import client
def call_llm(zero_shot_system_message1_1,user_input1_1, model="openai/gpt-oss-120b", temperature=0.2):
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "system", "content": zero_shot_system_message1_1},
                  {"role":"user", "content":user_input1_1}],
        temperature=temperature,
        
 )
    return response.choices[0].message.content
```

```python
import json
from prompts import (
    zero_shot_system_message1_1,
    user_input1_1
)

response = call_llm(
    zero_shot_system_message1_1,
    user_input1_1
)

print("RAW RESPONSE:\n")
print(response)

try:
    data = json.loads(response)
    

    print("\nVALID JSON OUTPUT:\n")
    print(json.dumps(data, indent=4))

    with open("zero_shot_outputs1_1.json", "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4, ensure_ascii=False)

    print("\nSaved Successfully")

except Exception as e:
    print("Error:", e)

```

SELF-CONSISTENCY CASE - 5.1

```python
from groq_client import client
from prompts import system_message, answers_template, claim_context

answers_prompt = [
    {
        "role": "system",
        "content": system_message
    },
    {
        "role": "user",
        "content": answers_template.format(
            context=claim_context,
            question="What is the final reimbursable amount?",
            num_answers=8
        )
    }
]

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=answers_prompt,
    temperature=0.7
)

self_consistency_answers = response.choices[0].message.content

print("STEP 1 OUTPUT:\n")
print(self_consistency_answers)
```

```python
from groq_client import client
from prompts import system_message, consistency_template

consistency_prompt = [
    {
        "role": "system",
        "content": system_message
    },
    {
        "role": "user",
        "content": consistency_template.format(
            question="What is the final reimbursable amount?",
            answers=self_consistency_answers
        )
    }
]

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=consistency_prompt,
    temperature=0
)

final_result = response.choices[0].message.content

print("\nFINAL OUTPUT:\n")
print(final_result)
```

```python
import json

output = {
    "step1_answers": self_consistency_answers,
    "final_output": final_result
}

with open("self_consistency_case5_1.json", "w", encoding="utf-8") as f:
    json.dump(output, f, indent=4, ensure_ascii=False)

print("Saved successfully")
```

SELF CONSISTENCY CASE - 5.2

```python
from groq_client import client
from prompts import system_message5_2, user_case5_2

runs = []

for i in range(5):
    response = client.chat.completions.create(
        model="openai/gpt-oss-120b",
        messages=[
            {"role": "system", "content": system_message5_2},
            {"role": "user", "content": user_case5_2}
        ],
        temperature=0
    )

    runs.append(response.choices[0].message.content)

print("STEP 1 RUNS:\n")
for r in runs:
    print(r)
```

```python
from groq_client import client
from prompts import system_message5_2,consistency_template5_2


consistency_prompt = [
    {
        "role": "system",
        "content": system_message5_2
    },
    {
        "role": "user",
        "content": consistency_template5_2.format(runs=runs)
    }
]

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=consistency_prompt,
    temperature=0
)

final_output = response.choices[0].message.content

print("\nFINAL OUTPUT:\n")
print(final_output)
```

```python
import json
try:
    parsed = json.loads(final_output)

    with open("self_consistency_case5_2.json", "w", encoding="utf-8") as f:
        json.dump(parsed, f, indent=4, ensure_ascii=False)

    print("Parsed JSON saved")

except Exception as e:
    print("Not valid JSON, saved raw instead:", e)
```
