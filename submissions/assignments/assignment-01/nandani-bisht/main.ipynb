### Self-Consistency

Case 5.1 - Reimbursement Amount Calculation

```python
prompt = f"{system_prompt_task9}\n\n{user_prompt_task9}"

def get_dict(parsed, key):
    if isinstance(parsed, list):
        for item in parsed:
            if isinstance(item, dict) and key in item:
                return item
        return parsed[-1] if parsed else {}
    return parsed

individual_answers = []
for _ in range(5):
    response = call_llm(prompt, temperature=0.7)
    parsed = get_dict(extract_json(response), "final_reimbursable_amount")
    individual_answers.append(parsed.get("final_reimbursable_amount", 0))

vote_counts = dict(Counter(str(v) for v in individual_answers))
final_amount = float(max(vote_counts, key=vote_counts.get))

result = {
    "individual_answers": individual_answers,
    "final_reimbursable_amount": final_amount,
    "consistency_count": vote_counts,
    "final_decision": "Approved" if final_amount > 0 else "Rejected",
    "reasoning_summary": "Five independent runs were executed and the most frequent final_reimbursable_amount was selected."
}
save_output("self_consistency1_outputs.json", result)
print(json.dumps(result, indent=2))
```

Case 5.2 - Security Risk Classification

```python
prompt = f"{system_prompt_task10}\n\n{user_prompt_task10}"

def get_dict(parsed, key):
    if isinstance(parsed, list):
        for item in parsed:
            if isinstance(item, dict) and key in item:
                return item
        return parsed[-1] if parsed else {}
    return parsed

run_results = []
for _ in range(5):
    response = call_llm(prompt, temperature=0.7)
    parsed = get_dict(extract_json(response), "final_risk_level")
    run_results.append(parsed.get("final_risk_level", ""))

risk_votes = dict(Counter(run_results))
final_risk = max(risk_votes, key=risk_votes.get)

result = {
    "runs": run_results,
    "risk_level_votes": risk_votes,
    "final_risk_level": final_risk,
    "disagreement_analysis": "Variation across runs typically stems from different interpretations of Germany as a known VPN country.",
    "final_reasoning_summary": "Five independent runs were aggregated. The majority vote determined the final risk level."
}
save_output("self_consistency2_outputs.json", result)
print(json.dumps(result, indent=2))
```
