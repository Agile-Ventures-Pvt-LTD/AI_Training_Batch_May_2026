## Self-Consistency

#### Case 5.1 - Self-Consistency for Complex Policy Interpretation

```python
import json
import os
from collections import Counter

from prompts import SELF_CONSISTENCY_POLICY_PROMPT
from groq_client import call_llm
from utils import extract_json


def main():

    individual_answers = []

    # Run model 5 times
    for i in range(5):

        response = call_llm(
            prompt=SELF_CONSISTENCY_POLICY_PROMPT,
            temperature=0.2
        )

        parsed = extract_json(response)

        amount = parsed.get("reimbursable_amount")

        individual_answers.append({
            "run": i + 1,
            "answer": parsed,
            "amount": amount
        })

    # Aggregate answers
    amounts = [
        item["amount"]
        for item in individual_answers
        if item["amount"] is not None
    ]

    counter = Counter(amounts)

    final_amount = counter.most_common(1)[0][0]

    final_output = {
        "individual_answers": individual_answers,
        "final_reimbursable_amount": final_amount,
        "consistency_count": dict(counter),
        "final_decision": f"Approved reimbursement amount: ${final_amount}",
        "reasoning_summary": (
            "Most runs consistently applied alcohol exclusion, "
            "international uplift, and same-day travel reimbursement rules."
        )
    }

    # Save output
    os.makedirs("outputs", exist_ok=True)

    with open(
        "outputs/5.1_self_consistency_policy_outputs.json",
        "w",
        encoding="utf-8"
    ) as f:
        json.dump(final_output, f, indent=4)
    
    print("SELF CONSISTENCY RESULT")

    print(json.dumps(final_output, indent=4))


if __name__ == "__main__":
    main()
```

#### Case 5.2 - Self-Consistency for Logical Deduction


```python
import json
import os
from collections import Counter

from prompts import SELF_CONSISTENCY_SECURITY_PROMPT
from groq_client import call_llm
from utils import extract_json


def main():

    runs = []

    # Run model 5 times independently
    for i in range(5):

        response = call_llm(
            prompt=SELF_CONSISTENCY_SECURITY_PROMPT,
            temperature=0.2
        )

        parsed = extract_json(response)

        risk_level = parsed.get("risk_level")

        runs.append({
            "run": i + 1,
            "response": parsed,
            "risk_level": risk_level
        })

    # Aggregate votes
    votes = [
        item["risk_level"]
        for item in runs
        if item["risk_level"]
    ]

    counter = Counter(votes)

    final_risk_level = counter.most_common(1)[0][0]

    # Analyze disagreement
    disagreement_analysis = ""

    if len(counter) == 1:
        disagreement_analysis = (
            "All runs consistently identified MEDIUM risk "
            "because the login occurred outside business hours "
            "and MFA failed once, while HIGH risk conditions "
            "were not satisfied."
        )
    else:
        disagreement_analysis = (
            "Some runs incorrectly treated Germany as a new country "
            "and escalated risk to HIGH or CRITICAL despite Germany "
            "being a known and VPN-approved country."
        )

    final_output = {
        "runs": runs,
        "risk_level_votes": dict(counter),
        "final_risk_level": final_risk_level,
        "disagreement_analysis": disagreement_analysis,
        "final_reasoning_summary": (
            "The final classification is MEDIUM risk because "
            "the login occurred outside business hours and "
            "there was one MFA failure. HIGH risk conditions "
            "were not met because Germany is not a new country."
        )
    }

    # Save output
    os.makedirs("outputs", exist_ok=True)

    with open(
        "outputs/5.2_self_consistency_security_outputs.json",
        "w",
        encoding="utf-8"
    ) as f:
        json.dump(final_output, f, indent=4)

    # Print output
    
    print("SELF CONSISTENCY SECURITY RESULT")

    print(json.dumps(final_output, indent=4))


if __name__ == "__main__":
    main()
```
