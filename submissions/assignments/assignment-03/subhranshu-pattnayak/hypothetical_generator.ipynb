## KEYS AND MODELS

````python
API_KEYS = [
    os.getenv("GROQ_API_KEY1"),
    os.getenv("GROQ_API_KEY2"),
    os.getenv("GROQ_API_KEY3"),
    os.getenv("GROQ_API_KEY4"),
    os.getenv("GROQ_API_KEY5"),
    os.getenv("GROQ_API_KEY6"),
    os.getenv("GROQ_API_KEY7"),
]

MODEL_FALLBACKS = [
    "openai/gpt-oss-120b",
    "openai/gpt-oss-20b",
    "openai/gpt-oss-safeguard-20b",
    "meta-llama/llama-prompt-guard-2-22m",
    "meta-llama/llama-prompt-guard-2-86m",
    "llama-3.3-70b-versatile",
    "llama-3.1-8b-instant",
    "qwen/qwen3-32b",
]

MAX_CONCURRENT_REQUESTS = 5
MAX_RETRIES = 5

OUTPUT_FILE = "data/hyde_questions.jsonl"
FAILED_FILE = "data/failed_batches.jsonl"
````

## CYCLE FUNCTION

````python
key_cycle = cycle(API_KEYS)

def get_client():
    key = next(key_cycle)
    return AsyncGroq(api_key=key)
````

## EXTRACT JSON FUNCTION

````python
def extract_json(content):

    content = content.strip()

    if "```" in content:
        content = (
            content
            .replace("```json", "")
            .replace("```", "")
            .replace("Here is the JSON output:", "")
            .strip()
        )

    start = content.find("[")
    end = content.rfind("]")

    if start != -1 and end != -1:
        content = content[start:end+1]

    return json.loads(content)
````

## SAVE RESULTS

````python
file_lock = asyncio.Lock()

async def save_result(records):

    async with file_lock:

        with open(
            OUTPUT_FILE,
            "a",
            encoding="utf-8"
        ) as f:

            for record in records:

                f.write(
                    json.dumps(
                        record,
                        ensure_ascii=False
                    )
                )

                f.write("\n")
````

## MESSAGES

````python
hypothetical_questions_system_message = """
You are a dataset generation assistant.

You will be given a document in python dictionary format where the dictionary will have 8 ids of chunks in 'ids' key and their respective chunks in 'documents' key.
Your task is to generate exactly 3 high-quality questions from a each chunk in the document.

Rules:
1. Generate questions answerable from the chunk alone.
2. Questions should cover different aspects of the chunk.
3. Avoid yes/no questions.
4. Avoid duplicate or highly similar questions.
5. Questions should be concise and natural.
6. Return ONLY valid JSON.
7. Do not include explanations, markdown, or extra text.
8. Questions from each should be paired with their respective chunk_id in a dictionary.
9. All dictionaries need to be put in a list and returned.

Output format for each chunk:

{
    "chunk_id": "<chunk_id>",
    "questions": [
        "question 1",
        "question 2",
        "question 3"
    ]
}

Output format for the whole response:
[
    {
        "chunk_id": "<chunk_id>",
        "questions": [
            "question 1",
            "question 2",
            "question 3"
        ]
    },
    {
        "chunk_id": "<chunk_id+1>",
        "questions": [
            "question 1",
            "question 2",
            "question 3"
        ]
    },
    {
        "chunk_id": "<chunk_id+2>",
        "questions": [
            "question 1",
            "question 2",
            "question 3"
        ]
    },
    and so on
]
"""

user_prompt = """
Batch:
{document}
"""
````

## GENERATION FUNCTION

````python
# This code snippet defines an asynchronous function `generate_questions` that takes `batch_data` as input. The function attempts to generate questions using different models specified in `MODEL_FALLBACKS` list. It makes multiple retries (up to `MAX_RETRIES`) in case of exceptions.
async def generate_questions(batch_data):

    last_exception = None

    for model in MODEL_FALLBACKS:

        for retry in range(MAX_RETRIES):

            try:

                client = get_client()
                response = await client.chat.completions.create(
                    model=model,
                    messages=[
                        {
                            "role": "system",
                            "content": hypothetical_questions_system_message
                        },
                        {
                            "role": "user",
                            "content": user_prompt.format(
                                document=batch_data
                            )
                        }
                    ],
                    temperature=0,
                )

                parsed = extract_json(
                    response.choices[0].message.content
                )

                return parsed

            except Exception as e:

                last_exception = e

                wait_time = min(
                    2 ** retry,
                    60
                )

                print(
                    f"{model} failed "
                    f"(retry={retry}) "
                    f"{str(e)}"
                )

                await asyncio.sleep(wait_time)

    raise last_exception
````

````python
# This line of code is creating a semaphore object using the `asyncio.Semaphore` class with a specified maximum number of concurrent requests. The `MAX_CONCURRENT_REQUESTS` variable is likely defined elsewhere in the code and represents the maximum number of allowed concurrent requests that can be made. The semaphore is used to control access to a shared resource or limit the number of simultaneous tasks that can access a particular resource.
semaphore = asyncio.Semaphore(
    MAX_CONCURRENT_REQUESTS
)
````

````python
import json
import os

processed_chunks = set()

# This code snippet is checking if a file specified by the variable `OUTPUT_FILE` exists. If the file exists, it opens the file in read mode with UTF-8 encoding. It then reads the file line by line, attempts to parse each line as JSON, and extracts the value associated with the key "chunk_id" from the JSON object. This extracted value is then added to a set called `processed_chunks`.
if os.path.exists(OUTPUT_FILE):

    with open(
        OUTPUT_FILE,
        "r",
        encoding="utf-8"
    ) as f:

        for line in f:

            try:

                item = json.loads(line)

                processed_chunks.add(
                    item["chunk_id"]
                )

            except Exception:
                pass

print(
    f"Loaded {len(processed_chunks)} processed chunks"
)
````

````python
# The code snippet defines an asynchronous function `process_batch` that processes a batch of data. Here's a breakdown of what the function does:
async def process_batch(
    batch_idx,
    batch_data
):

    batch_chunk_ids = set(
        batch_data["ids"]
    )

    if batch_chunk_ids.issubset(
        processed_chunks
    ):

        print(
            f"Batch {batch_idx} already exists. Skipping."
        )

        return

    async with semaphore:

        try:

            result = await generate_questions(
                batch_data
            )

            await save_result(
                result
            )

            for item in result:

                processed_chunks.add(
                    item["chunk_id"]
                )

            print(
                f"Batch {batch_idx} ✓"
            )

        except Exception as e:

            print(
                f"Batch {batch_idx} ✗"
            )

            with open(
                FAILED_FILE,
                "a",
                encoding="utf-8"
            ) as f:

                f.write(
                    json.dumps(
                        {
                            "batch_idx": batch_idx,
                            "error": str(e)
                        }
                    )
                )

                f.write("\n")
````

````python
# This code snippet defines an asynchronous function `run_all()` that processes batches of data using asyncio tasks. Here's a breakdown of what the code does:
RESUME_BATCH = 0
async def run_all():

    tasks = []

    for idx in range(
        RESUME_BATCH,
        len(batch)
    ):

        tasks.append(
            asyncio.create_task(
                process_batch(
                    idx,
                    batch[idx]
                )
            )
        )

    await asyncio.gather(*tasks)
````

````python
def generate_questions(
    batch_idx,
    batch_data
):

    print(
        f"Processing batch {batch_idx}"
    )

    response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {
                "role": "system",
                "content": hypothetical_questions_system_message
            },
            {
                "role": "user",
                "content": user_prompt.format(
                    document=batch_data
                )
            }
        ],
        temperature=0,
        top_p=0.7,
        max_tokens=2048,
        stream=False
    )

    return extract_json(
        response.choices[0].message.content
    )
````

## **CLEANING**

````python
import json
from collections import Counter

INPUT_FILE = "hyde_questions.jsonl"
OUTPUT_FILE = "hyde_questions_clean.json"

# Store unique records
unique_records = {}
duplicate_counter = Counter()

with open(INPUT_FILE, "r", encoding="utf-8") as f:

    for line_num, line in enumerate(f, start=1):

        try:
            record = json.loads(line)

            chunk_id = record["chunk_id"]

            duplicate_counter[chunk_id] += 1

            # Keep first occurrence only
            if chunk_id not in unique_records:
                unique_records[chunk_id] = record

        except Exception as e:
            print(
                f"Failed to parse line {line_num}: {e}"
            )

# Build ordered list
final_records = []

missing_chunks = []

for i in range(3336):

    chunk_id = f"text_{i}"

    if chunk_id in unique_records:

        final_records.append(
            unique_records[chunk_id]
        )

    else:

        missing_chunks.append(
            chunk_id
        )

# Save clean JSON
with open(
    OUTPUT_FILE,
    "w",
    encoding="utf-8"
) as f:

    json.dump(
        final_records,
        f,
        indent=4,
        ensure_ascii=False
    )

# Duplicate report
duplicates = {
    chunk_id: count
    for chunk_id, count in duplicate_counter.items()
    if count > 1
}

print("=" * 50)
print(f"Expected chunks: 3336")
print(f"Unique chunks found: {len(unique_records)}")
print(f"Clean records written: {len(final_records)}")
print(f"Missing chunks: {len(missing_chunks)}")
print(f"Duplicate chunk_ids: {len(duplicates)}")
print("=" * 50)

if duplicates:

    print("\nDuplicate chunk_ids:")
    for chunk_id, count in sorted(
        duplicates.items(),
        key=lambda x: int(x[0].split("_")[1])
    ):
        print(f"{chunk_id}: {count} occurrences")

if missing_chunks:

    print("\nMissing chunk_ids:")
    for chunk_id in missing_chunks:
        print(chunk_id)
````

````python
import traceback
import time

for batch_idx in batches_to_process:

    try:

        batch_data = batch[batch_idx]

        print("\n" + "=" * 60)
        print(f"STARTING BATCH {batch_idx}")

        print(
            f"Chunk IDs: "
            f"{batch_data['ids']}"
        )

        start_time = time.time()

        print(
            f"Sending request for "
            f"batch {batch_idx}..."
        )

        response = client.chat.completions.create(
            model=MODEL,
            messages=[
                {
                    "role": "system",
                    "content": hypothetical_questions_system_message
                },
                {
                    "role": "user",
                    "content": user_prompt.format(
                        document=batch_data
                    )
                }
            ],
            temperature=0
        )

        elapsed = (
            time.time()
            - start_time
        )

        print(
            f"Response received "
            f"({elapsed:.2f}s)"
        )

        try:

            parsed = extract_json(
                response.choices[0].message.content
            )

            print(
                f"JSON parsed successfully."
            )

            print(
                f"Records returned: "
                f"{len(parsed)}"
            )

        except Exception:

            print(
                f"JSON parsing failed "
                f"for batch {batch_idx}"
            )

            print(
                "\nRaw response:\n"
            )

            print(
                response.choices[0].message.content
            )

            raise

        try:

            save_result(parsed)

            print(
                f"Batch {batch_idx} saved."
            )

        except Exception:

            print(
                f"Save failed for "
                f"batch {batch_idx}"
            )

            raise

        print(
            f"BATCH {batch_idx} COMPLETE"
        )

    except Exception as e:

        print(
            f"\nBATCH {batch_idx} FAILED"
        )

        print(
            f"Error: {e}"
        )

        traceback.print_exc()

        with open(
            "manual_recovery_failures.log",
            "a",
            encoding="utf-8"
        ) as f:

            f.write(
                f"\nBatch {batch_idx}\n"
            )

            f.write(
                f"Error: {str(e)}\n"
            )

            f.write(
                traceback.format_exc()
            )

            f.write(
                "\n" + "=" * 60 + "\n"
            )

        continue
````

````python
def hyde_retrieve(
    query,
    k=10
):

    hyde_results = hyde_collection.query(
        query_texts=[query],
        n_results=k
    )

    parent_ids = hyde_results["ids"][0]

    original_docs = vectorstore_persisted.get(
        ids=parent_ids
    )

    return "\n---\n".join(original_docs['documents'])
````
