# Chunking process

```python
import time

batch_size = 300

for start_idx in range(0, len(tesla_10k_chunks), batch_size):

    batch_docs = tesla_10k_chunks[start_idx:start_idx + batch_size]

    batch_ids = [
        f"text_{i}"
        for i in range(start_idx, start_idx + len(batch_docs))
    ]

    vectorstore_persisted.add_documents(
        documents=batch_docs,
        ids=batch_ids
    )

    print(f"Processed {start_idx + len(batch_docs)} chunks")

    time.sleep(5)  
```

# Hypothetical Questions Generation

```python
hypothetical_questions_system_message = """
Generate a list of exactly 3 hypothetical questions that the document presented in the input could be used to answer.
Generate only a list of questions, each question in a new line.
Do not number the questions or use bullet points.
Do not mention anything before or after the list.
"""

user_message_template = """
<Document>
{document}
</Document>
"""
```

```python
hypothetical_questions = []
```

```python
documents
```

```python
for document in documents:

    try:
        response = llm.invoke(
            [
                ('system', hypothetical_questions_system_message),
                ('user', user_message_template.format(document=documents.page_content))
            ]
        )

        questions = response.content.strip()
    except Exception as e:
        questions = ""

    questions_list = questions.split("\n")

    for question in questions_list:

        questions_metadata = {
            'parent_chunk_id': document.id,
            'parent_collection': 'full_document_chunks'
        }

        hypothetical_questions.append(
            Document(
                page_content=question,
                metadata=questions_metadata
            )
        )
```

```python
hypothetical_questions_vectorstore = Chroma(
    collection_name="hypothetical_questions",
    collection_metadata={"hnsw:space": "cosine"},
    embedding_function=embedding,
    client=chromadb_client
)
```

```python
chromadb_client.list_collections()
```

```python
hypothetical_questions_vectorstore.add_documents(
    documents=hypothetical_questions
)
```

```python
retriever = hypothetical_questions_vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={'k': 5}
)
```

```python
user_query = "What should a board member ask about risks that could prevent Tesla from meeting production, delivery, or scaling expectations? "
```

```python
hypothetical_questions_retrieved = retriever.invoke(user_query)
```

```python
hypothetical_questions_retrieved
```

```python
retrived_documents = vectorstore_persisted.get(
    ids=list(set([d.metadata['parent_chunk_id'] for d in hypothetical_questions_retrieved]))
)
```

# Saving the generated questions in .json file

```python
hyde_record = {
    "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    "original_query": user_query,
    "hypothetical_answer": str(hypothetical_questions_retrieved)
}
```

```python
import json
import os

json_file = "hyde_history.json"

if os.path.exists(json_file):

    try:
        with open(json_file, "r") as f:
            history = json.load(f)

    except json.JSONDecodeError:
        history = []

else:
    history = []

history.append(hyde_record)

with open(json_file, "w") as f:
    json.dump(history, f, indent=4)

print("HyDE output saved.")
```
