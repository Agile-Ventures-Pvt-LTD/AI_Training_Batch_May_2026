```python
def generate_questions_batch(chunks):

    combined = ""

    for i,chunk in enumerate(chunks):

        combined += f"""

CHUNK {i+1}:
{chunk[:250]}
"""

    prompt = f"""
You are improving retrieval for a RAG system.

Read ALL chunks together.

Generate ONLY 3 high-quality hypothetical questions
that could retrieve this ENTIRE batch.

NOT 3 per chunk.

Return ONLY:

Q1
Q2
Q3

BATCH CONTENT:

{combined}
"""

    response = nvidia_client.chat.completions.create(

        model="meta/llama-3.1-8b-instruct",

        messages=[
            {
                "role":"user",
                "content":prompt
            }
        ],

        temperature=0
    )

    return response.choices[0].message.content.strip()
```

```python
def generate_hyde_query(user_query):

    prompt=f"""
Write a short hypothetical answer for retrieval.

Question:
{user_query}
"""

    response = client.chat.completions.create(

        model=model_name,

        messages=[
            {
                "role":"user",
                "content":prompt
            }
        ],

        temperature=0
    )

    return response.choices[0].message.content.strip()
```

```python
def respond_hyde(user_query):

    print("Original Query:")

    print(user_query)

    hyde_query = generate_hyde_query(
        user_query
    )

    print("\nGenerated HyDE Query:")

    print(hyde_query)

    relevant_document_chunks = retriever.invoke(
        hyde_query
    )

    print(
        f"\nRetrieved Chunks: {len(relevant_document_chunks)}"
    )

    context_list = []

    for i,doc in enumerate(
        relevant_document_chunks,
        1
    ):

        print(f"\n----- Chunk {i} -----")

        print(
            doc.page_content[:500]
        )

        context_list.append(
            doc.page_content
        )

    context_for_query = "\n---\n".join(
        context_list
    )

    prompt=[

        {
            'role':'system',
            'content':qna_system_message
        },

        {
            'role':'user',
            'content':qna_user_message_template.format(
                context=context_for_query,
                question=user_query
            )
        }
    ]

    response = client.chat.completions.create(

        model=model_name,

        messages=prompt,

        temperature=0
    )

    prediction = response.choices[0].message.content.strip()

    print("\nFINAL ANSWER:")

    print(prediction)

    return prediction
```

```python
respond_hyde(
    "What electric vehicle models does Tesla currently offer or plan to introduce for consumer and commercial markets?"
)
```
