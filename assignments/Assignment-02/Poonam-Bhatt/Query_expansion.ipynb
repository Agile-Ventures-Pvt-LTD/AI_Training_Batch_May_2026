```python
qna_system_message = """
You are an assistant to a financial services firm who answers user queries on annual reports.
User input will have the context required by you to answer user queries.
This context will be delimited by: <Context> and </Context>.
The context contains references to specific portions of a document relevant to the user query.

User queries will be delimited by: <Question> and </Question>.

Please answer user queries only using the context provided in the input.
Do not mention anything about the context in your final answer. Your response should only contain the answer to the question.

If the answer is not found in the context, respond "I don't know".
"""
```

```python
import json
import os
from datetime import datetime

QUERY_FILE = "query_expansion_log.json"
```

```python
def save_query_expansion(
    original_query,
    expanded_queries
):

    entry = {

        "timestamp":
            str(datetime.now()),

        "original_query":
            original_query,

        "expanded_queries":
            expanded_queries

    }

    if os.path.exists(
        QUERY_FILE
    ):

        with open(
            QUERY_FILE,
            "r"
        ) as f:

            data = json.load(f)

    else:

        data=[]

    data.append(entry)

    with open(
        QUERY_FILE,
        "w"
    ) as f:

        json.dump(
            data,
            f,
            indent=4
        )
```

```python
from groq import Groq

client = Groq(api_key=GROQ_API_KEY)
```

```python
def expand_query(query):

    prompt=f"""
You are improving retrieval for a Tesla RAG system.

Generate 3 semantically similar search queries.

Preserve EXACT meaning.

Question:
{query}

Return ONLY queries.
One per line.
"""

    response = client.chat.completions.create(

        model=model_name,

        messages=[
            {
                "role":"user",
                "content":prompt
            }
        ],

        temperature=0
    )

    queries = [

        q.strip()

        for q in response
        .choices[0]
        .message
        .content
        .split("\n")

        if q.strip()
    ]

    return list(set(queries))
```

```python



qna_system_message = """
You are an assistant to a financial services firm who answers user queries on annual reports.
User input will have the context required by you to answer user queries.
This context will be delimited by: <Context> and </Context>.
The context contains references to specific portions of a document relevant to the user query.

User queries will be delimited by: <Question> and </Question>.

Please answer user queries only using the context provided in the input.
Do not mention anything about the context in your final answer. Your response should only contain the answer to the question.

If the answer is not found in the context, respond "I don't know".
"""

qna_user_message_template = """
<Context>
Here are some documents that are relevant to the question mentioned below.
{context}
</Context>

<Question>
{question}
</Question>
"""

def respond(user_query):

    # STEP 1 — Generate expanded queries
    expanded_queries = expand_query(user_query)

    print("Original Query:")
    print(user_query)

    print("\nExpanded Queries:")
    for q in expanded_queries:
        print("-", q)

    # STEP 2 — Retrieve per expanded query
    all_docs = []

    for q in expanded_queries:

        docs = retriever.invoke(q)

        print(f"\nRetrieved {len(docs)} chunks for:")
        print(q)

        for i,doc in enumerate(docs,1):

            print(f"\nChunk {i}")

            print(doc.page_content[:300])

        all_docs.extend(docs)

    # STEP 3 — Deduplicate chunks
    unique_docs = list(
        {d.page_content:d for d in all_docs}.values()
    )

    print("\nUnique chunks:",len(unique_docs))

    # STEP 4 — Build context
    context_list = [
        d.page_content
        for d in unique_docs
    ]

    context_for_query = "\n---\n".join(context_list)

    # STEP 5 — Final prompt
    prompt = [

        {
            'role':'system',
            'content':qna_system_message
        },

        {
            'role':'user',
            'content':qna_user_message_template.format(
                context=context_for_query,
                question=user_query
            )
        }
    ]

    # STEP 6 — LLM answer
    response = client.chat.completions.create(
        model=model_name,
        messages=prompt,
        temperature=0
    )

    prediction = response.choices[0].message.content.strip()

    print("\nFINAL ANSWER:")
    print(prediction)

    return prediction
```
